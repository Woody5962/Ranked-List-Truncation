## 评估函数

```python
def cal_F1(ranked_list: list, query: str, k: int, N_D: int) -> float:
    """
    计算F1 score
    k: 截断到第k个，从1计数
    """
    count = sum([ranked_list[i] in gt[query] for i in range(k)])
    p_k = (count / k) if k != 0 else 0
    r_k = (count / N_D) if N_D != 0 else 0
    return (2 * p_k * r_k / (p_k + r_k)) if p_k + r_k != 0 else 0


def cal_DCG(ranked_list: list, query: str, k: int, N_D: int, penalty=-0.25, normalized=False) -> float:
    """
    计算DCG
    """
    value = 0
    for i in range(k): value += (1 / math.log(i + 2, 2)) if ranked_list[i] in gt[query] else (penalty / math.log(i + 2, 2))
    ideal_DCG = sum([1 / math.log(i + 2, 2) for i in range(k)]) if N_D >= k else sum([1 / math.log(i + 2, 2) for i in range(N_D)] + [penalty / math.log(i + 2, 2) for i in range(N_D, k)])
    return value if not normalized else value / ideal_DCG
```

## BM25与DRMM的greedy-k处理流程

```python
def ori2rt(dataset_name: str, original_data: dict) -> dict:
    rt_data = {}
    for key in original_data: 
        rt_data[key] = [original_data[key]['retrieved_documents'][i]['doc_id'] for i in range(300)] if dataset_name == 'BM25' else [original_data[key][i]['doc_id'] for i in range(300)]
    return rt_data


def dataset_prepare(dataset_name: str) -> list:
    train_dataset_list, test_dataset_list = [], []
    if dataset_name == 'BM25':
        for i in range(1, 6):
            with open('../data_prep/BM25_results/split_{}/BM25_train_s{}.pkl'.format(i, i), 'rb') as f: train_dataset_list.append(pickle.load(f))
            with open('../data_prep/BM25_results/split_{}/BM25_test_s{}.pkl'.format(i, i), 'rb') as f: test_dataset_list.append(pickle.load(f))
            train_dataset_list[-1] = ori2rt('BM25', train_dataset_list[-1])
            test_dataset_list[-1] = ori2rt('BM25', test_dataset_list[-1])
    else:
        for i in range(1, 6):
            with open('../data_prep/drmm_results/split_{}/drmm_train_s{}.pkl'.format(i, i), 'rb') as f: train_dataset_list.append(pickle.load(f))
            with open('../data_prep/drmm_results/split_{}/drmm_test_s{}.pkl'.format(i, i), 'rb') as f: test_dataset_list.append(pickle.load(f))
            train_dataset_list[-1] = ori2rt('DRMM', train_dataset_list[-1])
            test_dataset_list[-1] = ori2rt('DRMM', test_dataset_list[-1])
    return train_dataset_list, test_dataset_list


def train_scores(dataset: dict, gt: dict) -> tuple:
    F1_train, DCG_train, NDCG_train = [], [], []
    for key in dataset:
        per_k_F1, per_k_DCG, per_k_NDCG, N_D = [], [], [], sum(dataset[key][i] in gt[key] for i in range(300))
        for i in range(1, 300): 
            per_k_F1.append(cal_F1(dataset[key], key, i, N_D))
            per_k_DCG.append(cal_DCG(dataset[key], key, i, N_D))
            per_k_NDCG.append(cal_DCG(dataset[key], key, i, N_D, penalty=-0.5, normalized=True))
        F1_train.append(per_k_F1)
        DCG_train.append(per_k_DCG)
        NDCG_train.append(per_k_NDCG)
    F1_results, DCG_results, NDCG_results = list(np.mean(np.array(F1_train), axis=0)), list(np.mean(np.array(DCG_train), axis=0)), list(np.mean(np.array(NDCG_train), axis=0))
    scores = [max(F1_results), max(DCG_results), max(NDCG_results)]
    F1, DCG, NDCG = (scores[0], F1_results.index(scores[0])), (scores[1], DCG_results.index(scores[1])), (scores[2], NDCG_results.index(scores[2]))
    return F1, DCG, NDCG


def test_scores(dataset: dict, gt: dict, k: list) -> float:
    F1_test, DCG_test, NDCG_test = [], [], []
    for key in dataset:
        N_D = sum(dataset[key][i] in gt[key] for i in range(300))
        F1_test.append(cal_F1(dataset[key], key, min(k[0] + 1, 300), N_D))
        DCG_test.append(cal_DCG(dataset[key], key, min(k[1] + 1, 300), N_D))
        NDCG_test.append(cal_DCG(dataset[key], key, min(k[2] + 1, 300), N_D, penalty=-0.8, normalized=True))
    F1, DCG, NDCG = np.mean(F1_test), np.mean(DCG_test), np.mean(NDCG_test)
    return F1, DCG, NDCG


def k_fold(dataset_name: str) -> float:
    train_dataset_list, test_dataset_list = dataset_prepare(dataset_name)
    # 在训练集中得到贪心的结果列表
    F1_tuple_list, DCG_tuple_list, NDCG_tuple_list = [], [], []
    for dataset in train_dataset_list:
        result_tuples = train_scores(dataset, gt)
        F1_tuple_list.append(result_tuples[0])
        DCG_tuple_list.append(result_tuples[1])
        NDCG_tuple_list.append(result_tuples[2])
    F1_tuple_list, DCG_tuple_list, NDCG_tuple_list = sorted(F1_tuple_list, key=lambda x: x[0]), sorted(DCG_tuple_list, key=lambda x: x[0]), sorted(NDCG_tuple_list, key=lambda x: x[0])
    k = [F1_tuple_list[-1][1], DCG_tuple_list[-1][1], NDCG_tuple_list[-1][1]]
    # 在测试集上得到对应于贪心k的结果列表
    F1_score_list, DCG_score_list, NDCG_score_list = [], [], []
    for dataset in test_dataset_list:
        results = test_scores(dataset, gt, k)
        F1_score_list.append(results[0])
        DCG_score_list.append(results[1])
        NDCG_score_list.append(results[2])
    return np.mean(F1_score_list), np.mean(DCG_score_list), np.mean(NDCG_score_list)
```

## 经过召回优化的drmm-tks处理流程

```python
def ori2rl(original_data: dict) -> dict:
    rl_data = {}
    for key in original_data: 
        rl_data[key] = [original_data[key][i]['doc_id'] for i in range(300)]
    return rl_data


def dataset_prepare() -> list:
    train_dataset_list, test_dataset_list = [], []
    for i in range(1, 6):
        with open('../data_prep/my_results/drmm_tks_results/split_{}/drmm_tks_train_s{}.pkl'.format(i, i), 'rb') as f: train_dataset_list.append(pickle.load(f))
        with open('../data_prep/my_results/drmm_tks_results/split_{}/drmm_tks_test_s{}.pkl'.format(i, i), 'rb') as f: test_dataset_list.append(pickle.load(f))
        train_dataset_list[-1] = ori2rl(train_dataset_list[-1])
        test_dataset_list[-1] = ori2rl(test_dataset_list[-1])
    return train_dataset_list, test_dataset_list


def train_scores(dataset: dict, gt: dict) -> tuple:
    F1_train, DCG_train = [], []
    for key in dataset:
        per_k_F1, per_k_DCG, N_D = [], [], sum(dataset[key][i] in gt[key] for i in range(300))
        for i in range(300): 
            per_k_F1.append(cal_F1(dataset[key], key, i, N_D))
            per_k_DCG.append(cal_DCG(dataset[key], key, i, N_D))
        F1_train.append(per_k_F1)
        DCG_train.append(per_k_DCG)
    F1_results, DCG_results = list(np.mean(np.array(F1_train), axis=0)), list(np.mean(np.array(DCG_train), axis=0))
    scores = [max(F1_results), max(DCG_results)]
    F1, DCG = (scores[0], F1_results.index(scores[0])), (scores[1], DCG_results.index(scores[1]))
    return F1, DCG


def test_scores(dataset: dict, gt: dict, k: list) -> float:
    F1_test, DCG_test = [], []
    for key in dataset:
        N_D = sum(dataset[key][i] in gt[key] for i in range(300))
        F1_test.append(cal_F1(dataset[key], key, min(k[0] + 1, 300), N_D))
        DCG_test.append(cal_DCG(dataset[key], key, min(k[1] + 1, 300), N_D))
    F1, DCG = np.mean(F1_test), np.mean(DCG_test)
    return F1, DCG


def k_fold() -> float:
    train_dataset_list, test_dataset_list = dataset_prepare()
    # 在训练集中得到贪心的结果列表
    F1_tuple_list, DCG_tuple_list = [], []
    for dataset in train_dataset_list:
        result_tuples = train_scores(dataset, gt)
        F1_tuple_list.append(result_tuples[0])
        DCG_tuple_list.append(result_tuples[1])
    F1_tuple_list, DCG_tuple_list = sorted(F1_tuple_list, key=lambda x: x[0]), sorted(DCG_tuple_list, key=lambda x: x[0])
    k = [F1_tuple_list[-1][1], DCG_tuple_list[-1][1]]
    # 在测试集上得到对应于贪心k的结果列表
    F1_score_list, DCG_score_list = [], []
    for dataset in test_dataset_list:
        results = test_scores(dataset, gt, k)
        F1_score_list.append(results[0])
        DCG_score_list.append(results[1])
    return np.mean(F1_score_list), np.mean(DCG_score_list)
```

### 针对BM25的ranked list生成结果

```python
BM_F1_train, BM_DCG_train, BM_NDCG_train = [], [], []  # 三个矩阵
for key in BM25_train_list_s1:
    per_k_F1, per_k_DCG, per_k_NDCG, N_D = [], [], [], sum(BM25_train_list_s1[key][i] in gt[key] for i in range(300))
    for i in range(1, 300): 
        per_k_F1.append(cal_F1(BM25_train_list_s1[key], key, i, N_D))
        per_k_DCG.append(cal_DCG(BM25_train_list_s1[key], key, i, N_D))
        per_k_NDCG.append(cal_DCG(BM25_train_list_s1[key], key, i, N_D, penalty=-1, normalized=True))
    BM_F1_train.append(per_k_F1)
    BM_DCG_train.append(per_k_DCG)
    BM_NDCG_train.append(per_k_NDCG)

BM_F1_results, BM_DCG_results, BM_NDCG_results = np.mean(np.array(BM_F1_train), axis=0), np.mean(np.array(BM_DCG_train), axis=0), np.mean(np.array(BM_NDCG_train), axis=0)  # 训练集上针对每个k的结果
```
